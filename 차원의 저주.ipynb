> ## The Curse of Dimensionality

### Why High Dimensiomal Data Can Be So Troublesome??

---

엄청난 양의 데이터에 직면했을 때, 우리는 차원 축소 알고리즘을 사용하여 데이터를 요점에 도달하도록 만들 수 있다.
- 이를테면 다음과 같은 알고리즘이 있을 수 있다. 
  - PCA
  - non-negative matrix factorization
  - linear discriminator amnalysis

> #### 그렇다면 우리는 왜 차원 축소 알고리즘이 필요할까?

먼저 언제 고차원의 data가 발생하며 왜 그것이 문제로 이어지는가?

- 너무 많은 feature의 data
  - -> 예를 들어 100,000개의 potential lever가 있다면 즉, 100,000개의 feature가 있다면 100,000개의 잠재적 손상 변수가 손상 lever로서 작동할 위험이 도사리고 있는 것.

- 오늘날 빅데이터가 가져올 잠재적 issues
  - 1) 관측치 수 < feature 수
    - 과적합의 위험도 $\uparrow$
  - 2) feature가 너무 많으면 군집화의 장애물이 될 수 있음
    - 많은 feature, 즉 고차원의 data는 데이터 집합의 모든 관측치가 타 관측치와 동일하게 해석될 수 있음
      - Euclidean distance와 같은 방법은 거리를 이용하기 때문에 특히나 큰 문제로 이어질 수 있음
      - 거리가 대략 동일하다면 모든 관측치는 동일하게 판별될 것이고, 유의미한 군집화를 수행할 수 없음

> #### Dimensionality Reduction to the Rescue


- 차원 축소 알고리즘은 data feature의 근본적인 경향을 찾는 것으로 시작
- 사탕 dataset을 살펴보면, 근본적인 경향은 주요 색인 빨강과 파랑이 될 것이다.
- data science에서 feature에 숨겨진 이런 근본적인 경향은 latent feature라 부른다. 

- 그리고 다음 단계로서, 알고리즘은 색의 노출도, (여기선 latent features를 의미하는 것 같음)에 대해 다른 features의 재추정과 재기술을 수행한다.

- latent features인 exposures의 관점에서 재기술, 재추정된 feature 분류 그래프는 다음과 같다

- 아래 그림은 latent features로서 작용하는 exposures(노출도)의 관점에서 사탕을 plot한 것이다. 

- latent features로 하여금 두 군집으로 분류할 수 있었음
- 마침내 우리는 두 군집으로 분류돼 있는 data set을 기반으로 각 군집의 사탕들을 count할 수 있으며 이는 앞으로의 예측 기반이 될 것

$<Summary>$

- 1. 사탕이 주어지면, 그것의 색을 기록하라
- 2. 그것의 색을 빨강과 파랑의 노출도로 표현하라, 다시 말해서 latent features의 관점에서 색을 재표현하라
- 3. 분류할 새 사탕의 latent features를 이용함으로써 빨강 사탕 군집과 파랑 사탕 군집 중 어느 군집에 유사한지 결정하며 이떄 Euclidean distance와 같은 거리 측정 방법을 사용한다.
- 4. 우리의 모델이 new 사탕을 빨강 군집으로 분류한다면, 우리는 해당 new 사탕을 spicy하다고 생각할 것이다.

$<Conclusion>$

- 고차원의 data set은 특정 머신러닝 기법을 혼동시킬 수 있음
- 이때, 차원 축소를 이용할 수 있음

- 좀 더 많은 noise가 결부된 실생활 data에서 차원축소로 하여금 강력한 모델을 얻는 효과 대신 비교적 낮아진 해석 능력을 얻을 수 있다. 

> 차원의 저주를 해결할 도구가 있긴 하지만, 이 도구를 사용함으로써 따르는 비용에 대해서도 고려해보아야 한다.

> ### additional info

- 차원의 저주?
  - 차원이 증가하면서 학습데이터 수가 차원 수보다 적어져서 성능이 저하되는 현상을 일컫는다.
  - 차원이 증가할수록 변수(즉, feature)가 증가하고, 개별 차원내에서 학습할 데이터 수가 적어진다.
 > 이때 주의할 점은 변수가 증가한다고 반드시 차원의 저주가 발생하는 것은 아니다. 
 
 > 관측치보다 변수 수가, 즉 feature수가 많아지는 경우에 차원의 저주 문제가 발생하는 것이다.

그림에서 알 수 있듯, 차원이 증가(1$\to$2$\to$3)할수록 빈 공간이 많아지는 것을 확인할 수 있다. 

같은 데이터임에도 불구하고 1차원에서는 데이터 density가 촘촘했던 것이 2차원, 3차원으로 갈수록 점점 데이터간 거리가 멀어진다.

이렇게 차원이 증가하면 빈 공간이 생기게 되는데 빈 공간은 컴퓨터에서 0으로 채워진 공간, 즉 정보가 없는 공간이기 때문에 빈 공간이 많을수록 학습 시켰을 때 모델 성능이 저하될 수밖에 없다.

고차원 공간은 이렇게 빈 공간이 많아 훈련 데이터가 서로 멀리 떨어져 있으며, 새로운 샘플도 훈련 샘플과 멀리 떨어져 있을 가능성이 높다. 따라서 예측을 위해서는 외삽법(=보외법)을 많이 사용해야 하는데 이때 과적합의 문제가 따를 수 있다. 

---

- 외삽법(=보외법)?
  - 이전의 경험에 비추어 보다 과학적인 맥락에서는 이전의 실험으로부터 얻은 데이터들에 비추어, 아직 경험/실험하지 못한 경우를 예측해보는 기법
  - 어디까지나 엄밀한 추론은 아님에 유의
  - 이용가능한 자료의 범위가 한정되어 있어 그 범위 이상의 값을 구할 수 없을 때, 관측된 값을 이용해 한계점 이상의 값을 추정하는 것을 의미
  - 예를 들어, 아래 그림에서 A~B까지의 data만 주어져 있을 때 P값을 이용해 Q값을 예측하는 것이다. 당연히 예측 정확도는 하락한다.

- 이런 차원의 저주 문제에 치명적인 것이 바로 KNN이다.

- 차원이 커질수록 주변 이웃들이 멀어지니 예측에 사용할 포인트 값이 한참 떨어져 있어 모델 성능도 나빠지고 설명력도 좋지 않게 된다.

> ### 차원의 저주 해결?

- 훈련 샘플의 밀도가 충분히 높아질 때까지 데이터를 모아서 훈련 세트의 크기를 키우는 것
  - 하지만 일정 밀도에 도달하기 위해 필요한 훈련 샘플 수는 차원의 수가 커짐에 따라 기하급수적으로 늘어난다.
  - 이 방법으로는 해결하기 어렵다는 뜻
  - 어렵다가 아니라 불가능하다

- 따라서 현실적인 방법으로서 PCA와 같은 차원축소 기법을 이용하여 차원의 저주 문제를 해결한다
- feature 수를 크게 줄여 차원 축소를 하면 훈련속도가 빨라지고 시각화 하기에도 좋다.
  - 일부 정보가 유실될 수 있으며 성능도 조금 나빠질 수 있다. 
  - 무작정 차원 축소를 하기보다는 원본 데이터로 훈련해서 시간이 오래 걸리는지 확인하고 차원 축소 필요 여부를 결정할 필요가 있다. 

---

> 무조건 feature의 수가 증가한다고 해서 차원의 저주 문제가 있는 것이 아니라 관측치 수보다 feature의 수가 많아지면 발생한다. 

    예를 들면, 관측치 개수는 200개인데 변수는 7000개

근데 feature수가 7,000개일수가 있나?

어떤 상황일까?
- 1차원에 늘어져있던 data들을 2차원으로 표현, 혹은 3차원으로 표현하면 자연스레 빈 공간이 발생할 것이고, 이처럼 차원이 증가하며 표현함에 따라 빈 공간이 생기는 것을 차원의 저주임!
- 이때 빈공간이 계속 생기니, 관측치 수가 feature수보다 작을 수밖에 없는 것

- 차원의 저주 문제에 치명적인 알고리즘이 KNN이다.
- KNN 알고리즘은 거리 기반 자신과 가장 가까운 이웃 K개를 보고 label 값을 정하게 되는데, 차원이 커질수록 내 주변의 이웃이 점점 멀어지게 됨에 따라 KNN의 성능은 저하된다. 
- 따라서 KNN을 쓸 때, 너무 큰 차원이면 되도록 다른 알고리즘을 사용하거나 차원을 줄이는 방법으로 데이터를 한 번 정제해야 한다. (전처리)

---

> 미국 출신 소프트웨어 엔지니어 벤 딕슨이 제시하는 차원의 저주 해결 method

목적에 적합한 샘플 수만 남기고 나머지를 없앤다면 데이터 사용량을 줄이는 한편 머신러닝 모델은 더욱 단순하고 효율적으로 구축할 수 있다. 

이같은 과정을 차원축소라 부른다

1) 차원축소의 첫 번째 방법 : `Feature Selection`

- 요점만 분류
- 대상 feature와 가장 관련성이 높은 형상의 하위 집합을 식별해 제외시킨다.
- 예를 들어 산점도나 히트맵을 사용해 여러 형상의 공분산을 시각화하여 두 feature가 높은 상관관계를 띈다면 둘 다 대상 변수에 유사한 영향을 미치기 때문에 머신러닝 모델에 모두 포함시킬 필요가 없다.
> 둘중 하나를 제거한다

2) 차원축소의 두 번째 방법 : Feature Extraction

- 개별 기능을 제거할 수 없는 경우도 있다.
- Feature Extraction, 즉 투영기술을 통해 여러 형상을 저차원 공간으로 압축해 모델을 단순화한다.
  - PCA가 있다.
    - 데이터의 특성을 가장 잘 표현하는 주성분을 추출해 데이터 양을 줄이는 방법이다.

capm(자본자산 가격결정 모형)한계점
- 자산의 수익률을 결정하는 공통요인이 시장포트폴리오 수익률(시장 위험?) 하나라고 보는데, 자산의 수익률에 영향을 미치는 공통요인이 여러가지가 있을 수 있다.
- 자본자산 가격경정 모델은 개별종목의 총위험을 시장에 연관되어 나타나는 위험(체계적 위험)과 시장과 상관없이 나타나는 위험(비체계적 위험)으로 분류
- 그리고 시장과 상관없이 나타나는 위험은 분산투자에 의해 제거될 수 있다고 본다. 
  - 따라서 체계적 위험에서 보상받을 수 있는 방법은 시장과 관련된 베타 계수(시장 위험)뿐이다. 이런 의미에서 모든 투자자는 동시에 동일한 내용의 정보를 입수할 수 있다는 효율적 시장가설을 전제로 하고 있으며 어떤 분석에 의해서도 증권시장을 상회하는 것보다 저가의 증권을 계속해서 찾아낸다는 것은 곤란하다고 한다

APT의 가정
1) 자산의 수익률은 `여러 개`의 `서로 독립적`인 `공통요인`에 따라 결정된다.
  - $\star$$\star$$\star$ 이게 핵심 $\star$$\star$$\star$
  - 자산 또는 증권의 수익률은 여러 요인들의 선형함수로 표시할 수 있다
2) 투자자들은 위험회피형이며, 더욱 더 많은 부를 선호한다. 
3) 투자자들은 공통요인과 자산수익률의 확률분포에 대하여 동질적으로 기대한다. 
4) 자본시장은 완전시장이다.

- 완전 시장에서 차익거래이익을 극대화하기 위해 행동한다고 할 경우, 더 이상의 차익거래이익을 얻을 수 없는 균형상태에서는 모든 자산의 기대 수익률이 각 공통요인에 대한 민감도와 선형관계를 갖고 결정된다는 이론이다. 

APT에 의한 수익률은 아래와 같이 3가지로 구성된다.
1) 기대수익률
2) 공통요인의 예상하지 못한 변동으로 인한 수익률 변동
3) 개별자산의 고유요인으로 발생하는 수익률 변동

> 기존의 CAPM에서는 시장 리스크만을 고려했지만 APT에서는 다양한 요인들이 리스크의 원천이 될 수 있다. 

> APT는 위험 요인만을 찾으면 된다.

- 이 이론은 위험요인이 무엇인지에 대해서는 설명하지 않고 다만 일물일가의 법칙과 통계적 경향을 이용해 자산의 가격을 설명한다.
- 이를테면, 바나나 선물의 수익률이 주식 수익률을 잘 설명하는 것이 그레인저 인과관계등을 통해 통계적으로 확인된다면 바나나 선물의 수익률은 위험요인이 되는 것이다.

- API 서버 만들기
- 모델 관리
- 데이터 셋 관리
- ML 인프라 관리

이런 부분을 MLOps로 통칭, 굉장히 큰 개념이다

$\star$$\star$$\star$ 나는 mlops를 공부하겠다!는 사실 불가능까진 아니지만 상당히 현실적인 얘기는 아님, 그 범위가 상당히 방대하고, 도메인별로 mlops의 세세한 부분이 매우 상이할 것

$\star$$\star$$\star$ 따라서 요구되는 로드맵 : 일단 앞으로 mlops의 수요는 분명할 것이니, mlops에 대해선 장기적 관점에서 봤을 때 이런 업무를 하고 싶다는 개념 하에서, 현재 나의 역량에 맞는 방향으로 먼저 취업하고, 거기서 경험을 쌓아가며 이때 mlops로 어떻게 발전시킬 수 있을지에 대해 고민하는 것이 더 현실적일 것.

- Thus, a should get 175, which is their SV; since b’s SV is 425, they should get 425. How did we compute the SV’s? Let us take a’s case:
  - 사원 A의 경우에 대해서 SV's를 계산해보자
- We first find how many times a appears in a “team arrangement” (coalition in GT)
  - 팀 연합의 관점에서 a가 몇 번이나 나타나는가?
- Here, they appear twice — once by themselves, S₁ = {a}; and then with b, S₂ = {a, b}
  - 이 경우에선 a가 나타날 수 있는 경우의 수는 2가지이다.
  - a 혼자 일하거나 $\to$ S₁ = {a}
  - b와 같이 일하는 경우 $\to$ S₂ = {a, b}
- For each such coalition Sᵢ
  - 이러한 두 연합에 대해 살펴보자
- We find the value or contribution of that coalition: v(Sᵢ)
  - 각 연합에 대한 기여도 또는 가치에 대해 알아보고 이를 v(Sᵢ)로 표현하자
- Next, we remove a from the collation Sᵢ while retaining everybody else, and find the value of the truncated coalition Sᵢ’ (= Sᵢ -{a} = Sᵢ -a): v(Sᵢ’)
  - 그 다음, 각 연합에서 a를 제거한다.
    - 현재 우리는 a의 SV's를 계산하고 있음
  - 그리고 a가 제거됐을 때의 값을 구함
- ### $\star$$\star$$\star$ We then find the marginal value of a for Sᵢ as the difference of the above two values: dvᵢ = v(Sᵢ) — v(Sᵢ’)
- ### $\star$$\star$$\star$ We take the mean of all the marginal values generated as above.

> ## *A simple measure of the importance of the independent variables*

    It tells us that, for a unit change in xⱼ, y changes by βⱼ units while all other variables are held fixed. Therefore, the absolute or squared value of the regression coefficient βⱼ is a measure of the effect on y of a change in the input xⱼ. The greater is this value, the greater is the influence of xⱼ on y. Hence the size of this coefficient serves as a metric for the relative importance of xⱼ.

    An issue with this metric is that it is sensitive the scale or units of measurement of the input variables.

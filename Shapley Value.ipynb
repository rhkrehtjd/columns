> # `*Elements of Shapley Value Theory*`

    게임이론에 근간함
    게임이론? 어떤 활동이나 행위를 할 때 서로 영향을 미치는 상황에서 어떤 의사결정이나 행동을 하는지(결국 자신의 최대 이익에 부합하는 행동을 추구할 것)에 대해 이론화 한 것
    
    즉, shapley Value란 게임이론을 바탕으로 게임에서 각 플레이어의 기여도를 계산하는 것
    데이터 분석 각도에서 보자면, 하나의 특성에 대한 중요도를 알기 위해 여러 특성들의 조합을 구성하고 해당 특성의 유무에 따른 평균적인 변화를 통해 얻어낸 값
    
    게임이론을 바탕으로 여러 변수 중 한가지 변수가 결과에 미치는 중요도를 판단하기 위해 도입한 값
    
    팀의 성과를 팀 구성원 간에 공정하게 배분하는 것에 관한 것

> Shapley value는 모든 가능한 조합에 대해서 하나의 특성의 기여도를 종합적으로 합한 값이다.

> Shapley value는 모든 가능한 연합에 대한 모든 한계 기여도(marginal contributions)의 평균

> 한계 기여도? 해당 특성이 기여할 수 있는 한계값

> 각 연합에 대해서 cat-banned가 포함된 연합과 포함되지 않은 연합의 아파트 예측 가격을 계산하고 한계 기여도를 계산하여 차이를 구한다. 

> Shapley value는 한계 기여도의 (가중) 평균이다. 기계학습에서 예측치를 구하기위해 연합에 포함되지 않은 특성값은 아파트 데이터셋에서 무작위로 추출해서 대체한다.

- Thus, a should get 175, which is their SV; since b’s SV is 425, they should get 425. How did we compute the SV’s? Let us take a’s case:
  - 사원 A의 경우에 대해서 SV's를 계산해보자
- We first find how many times a appears in a “team arrangement” (coalition in GT)
  - 팀 연합의 관점에서 a가 몇 번이나 나타나는가?
- Here, they appear twice — once by themselves, S₁ = {a}; and then with b, S₂ = {a, b}
  - 이 경우에선 a가 나타날 수 있는 경우의 수는 2가지이다.
  - a 혼자 일하거나 $\to$ S₁ = {a}
  - b와 같이 일하는 경우 $\to$ S₂ = {a, b}
- For each such coalition Sᵢ
  - 이러한 두 연합에 대해 살펴보자
- We find the value or contribution of that coalition: v(Sᵢ)
  - 각 연합에 대한 기여도 또는 가치에 대해 알아보고 이를 v(Sᵢ)로 표현하자
- Next, we remove a from the collation Sᵢ while retaining everybody else, and find the value of the truncated coalition Sᵢ’ (= Sᵢ -{a} = Sᵢ -a): v(Sᵢ’)
  - 그 다음, 각 연합에서 a를 제거한다.
    - 현재 우리는 a의 SV's를 계산하고 있음
  - 그리고 a가 제거됐을 때의 값을 구함
- ### $\star$$\star$$\star$ We then find the marginal value of a for Sᵢ as the difference of the above two values: dvᵢ = v(Sᵢ) — v(Sᵢ’)
- ### $\star$$\star$$\star$ We take the mean of all the marginal values generated as above.

# The SV Formula

- n명의 플레이어가 있다. 
- n명 중 차출하여 구성한 조합들의 결과는 v()로 표현한다. 
  - 전체 플레이어의 수는 n이므로 가능한 부분집합의 개수는 2^n 이다. 
- 이전의 예시와 같이 S-i는 S와 동일 부분집합이며 단지 플레이어 i만 제외한 집합이다.

> 따라서 플레이어 i의 v()는 다음과 같다.

The formula expresses the SV of the player i as a weighted sum of the terms of the form [v(S) — v(S-i)]. 

Each such term is the marginal contribution of the player i to the coalition S.

이때 s는 부분집합에 속해있는 플레이어의 수이다.

> ### SV and Machine Learning

    dataset에서의 features = 게임 이론에서의 Players
    model output = Value generated
    
    the features cooperate to produce the model output
    
    each “row” of the data can be thought as a one particular coalition
    : dataset에서 각 행은 하나의 부분집합으로 생각할 수 있음
    
    with the SV apparatus, we can ascribe (more specifically, we can estimate) how much each feature might have contributed to the decision made by the model for that particular row — this is simply the SV of the feature
    : SV를 통해 featrue가 해당 row에서 얼마나 기여하고 있는지 알 수 있음, 구체적으로 말하면 추정할 수 있음

    This row-wise analysis can be considered local; it can also be extended to a global analysis to determine the contribution of a given feature to the decision over the entire dataset. 
    : 이러한 행별로 진행되는 분석은 local적인 분석.
    : 전체 데이터 세트의 결정에 대한 주어진 기능의 기여도를 결정하기 위해 전역 분석으로 확장될 수 있다.
